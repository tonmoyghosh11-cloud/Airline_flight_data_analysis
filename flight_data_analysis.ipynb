# Flight Data Analysis
This notebook uses **NumPy, Pandas, Matplotlib, and Seaborn** to explore the airlines flights dataset.

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For inline plotting
%matplotlib inline
```

```python

# Load the dataset
df = pd.read_csv("airlines_flights_data.csv")

# Show first 5 rows
df.head()
```

```python

# Dataset info
df.info()

# Summary statistics
df.describe(include='all')
```

```python

# Check missing values
df.isnull().sum()
```

```python

# Example with NumPy: calculate mean of numerical columns
np.mean(df.select_dtypes(include=[np.number]), axis=0)
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['price'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of Flight Price")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python

# Countplot for Airlines
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='airline', order=df['airline'].value_counts().index, palette="viridis")
plt.xticks(rotation=90)
plt.title("Number of Flights per airline")
plt.show()
```

```python

# Scatterplot example (Distance vs Price if available)
if 'Price' in df.columns:
    plt.figure(figsize=(8,6))
    sns.scatterplot(data=df, x='Distance', y='Price', hue='Airline')
    plt.title("Flight Distance vs Price")
    plt.show()
```

```python

# Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
